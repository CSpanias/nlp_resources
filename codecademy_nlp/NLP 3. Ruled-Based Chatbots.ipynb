```python
import re
import random

class PersonalBot:
    
    # possible responses to the question "what can I help you with?" and "Is there anything else I can help you with?"
    exit_commands = ("nothing", "don't", "stop", "no", "nope", "nothing",
                          "quit", "pause", "exit", "goodbye", "bye", "later")
    

    def __init__(self):
        self.matching_phrases = {
            'contact info': [r'.*contact.*', r'.*mail.*'],
            'resume': [r'.*resume.*', r'.*[Cc][Vv].*'],
            'user tel': [r'.*tel.*num.* (\d+)'],
        }
        # greet user and convert the answer to lower case letter
        self.name = input("Hi, my name is Charalampos Spanias. Welcome to my personal website. " 
        "What is your first name and last name? ").lower()
        
        

    def welcome(self):
        # a call to self.name
        self.name
        
        will_help = input(f"Ok {self.name}, what can I help you with? ").lower() 
        
        # if the word is in exit commands list exit the dialog
        if will_help in self.exit_commands:
          print("Ok, have a great day!")
          return
    
        # a call to the handle_conversation method
        self.handle_conversation(will_help)

        
    def handle_conversation(self, reply):
        
        while not self.make_exit(reply):
            reply = self.match_reply(reply)
          
        
    def make_exit(self, reply):
        
        for exit_command in self.exit_commands:
            if exit_command in reply:
                print("Ok, have a great day!")
                return True
            
        return False
    
    
    def match_reply(self, reply):
        for key, values in self.matching_phrases.items():
            for regex_pattern in values:
                found_match = re.match(regex_pattern, reply)
                
                if found_match and key == 'contact info':
                    return self.contact_info()
                elif found_match and key == 'resume':
                    return self.resume()
                elif found_match and key == 'user tel':
                    return self.user_tel(found_match.groups()[0])
                
        return input("I did not understand you. Can you please " 
        "ask your question differently?")
    
    
    def contact_info(self):
        return input("""My email address is: spaniascharalampos@gmail.com. 
        You can also contact me via my LinkedIn page. 
        I am looking forward to speaking with you!
        Is there anything else that I can help you with? """)
        
        
    def resume(self):
        return input("""You can find my resume here. You can also visit my LinkedIn and GitHub pages.
            Feel free to send me a message if you need additional information.
            Is there anything else that I can help you with? """)
        
    
    def user_tel(self, tel_number=None):
        return input(f"""Your telephone number: {tel_number} is saved and I would do my best to contact you asap. 
                     Is there anything else that I can help you with? """)
        

# Create a SupportBot instance
PersonalConversation = PersonalBot()
# Call the .welcome() method
PersonalConversation.welcome()
```
