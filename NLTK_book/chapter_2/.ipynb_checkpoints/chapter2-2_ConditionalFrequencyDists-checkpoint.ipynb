A CFD is a **collection of frequency distributions**, each one for a different "condition". The condition will often be the **category** of the text.

The figure below depicts a fragment of a conditional frequency distribution having just two conditions, one for news text and one for romance text.



<a name="CondEv"></a>
### 2.2.1 Conditions and Events
A **FreqDist counts observable events**, such as the appearance of words in a text. <br>

`text = ['The', 'Fulton', 'County', 'Grand', 'Jury', 'said']`

A **CDF** needs to **pair each event with a condition**. So instead of processing a sequence of words, we have to process a sequence of pairs. <br>

`pairs = [('news', 'The'), ('news', 'Fulton'), ('news', 'County')]`

Each pair has the form `(condition, event)`. 

If we were processing the entire **Brown Corpus** by genre there would be **15 conditions (one per genre)**, and **1,161,192 events (one per word)**.

<a name="CountGenre"></a>
### 2.2.2 Counting Words by Genre

`FreqDist(a_list)` a **simple list** as input

`ConditionalFreqDist(a_list_of_pairs)` a **list of pairs** as input



```python
from nltk.corpus import brown

# create a list of pairs
genre_word = [(genre, word)
             # assign the 2 genres to the temp var genre
             for genre in ["news", "romance"]
             # assign the words of each genre to the temp var word
             for word in brown.words(categories=genre)]

# find the length of the list of pairs
print("The list genre_word has {} pairs.".format(len(genre_word)))
```

As we can see below, pairs at the beginning of the list `genre_word` will be of the form `('news', word)`, while those at the end will be of the form `('romance', word)`.

```python
print("The first 5 pairs of genre_word are:\n{}\n".format(genre_word[:5]))

print("The last 5 pairs of genre_word are:\n{}".format(genre_word[-5:]))
```

We can now use our **list of pairs** to create a CFD.

```python
# create a CFD for our list of pairs
cfd_brown_genres = nltk.ConditionalFreqDist(genre_word)

# check the number of conditions
print(cfd_brown_genres)

# check the specific conditions
print("\nThe conditions of our CFD are: {}.".format(cfd_brown_genres.conditions()))
```

We can access the individual FreqDist.

```python
# check news condition
print("cfd_brown_genres[\"news\"] is a: {}.\n".format(cfd_brown_genres["news"]))

# check romance condition
print("cfd_brown_genres[\"romance\"] is a: {}.".format(cfd_brown_genres["romance"]))
```

### 2.2.3 Plotting and Tabulating Distributions
A CFD provides some useful methods for **tabulation** and **plotting**.

```python
# import the inaugural corpora
from nltk.corpus import inaugural

# create a CFD
cfd = nltk.ConditionalFreqDist(
    # create the pair of interest
    (target, fileid[:4])
    # for every file in inaugural corpora
    for fileid in inaugural.fileids()
    # for every word in that file
    for w in inaugural.words(fileid)
    # assign a list of words to target
    for target in ["america", "citizen"]
    # lower-case the word and check if it matches target
    if w.lower().startswith(target))

# generate CFD
cfd.plot();              
```

```python
# import udhr corpora
from nltk.corpus import udhr

# create a list with the languages of interest
languages = ['Chickasaw', 'English', 'German_Deutsch',
            'Greenlandic_Inuktikut', 'Hungarian_Magyar', 'Ibibio_Efik']

# create CFD
cfd = nltk.ConditionalFreqDist(
    (lang, len(word))
    for lang in languages
    for word in udhr.words(lang + '-Latin1')
)

# generate CFD plot
cfd.plot();
```

In the `plot()` and `tabulate()` methods, we can optionally **specify which conditions to display** with a `conditions= parameter`. When we omit it, we get all the conditions. 

Similarly, we can **limit the samples to display** with a `samples= parameter`. This makes it possible to load a large quantity of data into a conditional frequency distribution, and then to explore it by plotting or tabulating selected conditions and samples. 

It also gives us full control over the order of conditions and samples in any displays. 

For example, we can tabulate the **cumulative frequency** data just for **two languages**, and for **words less than 10 characters long**, as shown below.

We **interpret the last cell** to mean that 995 and 2074 words of the Chickasaw and Ibibio_Efik text, respectively, have 9 or fewer letters.

```python
# create CFD
cfd.tabulate(conditions=['Chickasaw', 'Ibibio_Efik'],
             samples=range(10), cumulative=True)
```

We can find out which days of the week are most "newsworthy" and which are most "romantic" using the brown corpora.

```python
from nltk.corpus import brown

# create a list with days
days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

# create a list of pairs
genre_word = [(genre, word)
             # assign the 2 genres to the temp var genre
             for genre in ["news", "romance"]
             # assign the words of each genre to the temp var word
             for word in brown.words(categories=genre)]

# create a CFD passing the list of pairs
cfd_days = nltk.ConditionalFreqDist(genre_word)

# create a tabulated format of the data
cfd_days.tabulate(samples=days, cumulative=True)
```

You may have noticed that the **multi-line expressions** we have been using with CFDs look like **list comprehensions, but without the brackets**.

In general, when we use a list comprehension as a parameter to a function, like set([w.lower() for w in t]), we are permitted to omit the square brackets and just write: set(w.lower() for w in t).

There are called [generator expressions](https://www.python.org/dev/peps/pep-0289/).

### 2.2.4 Generating Random Text with Bigrams

We can use a CDF to create a **table of bigrams**.

`bigrams()` takes a list of words and builds a list of consecutive word pairs. 

*Remember that, in order to see the result and not a cryptic "generator object", we need to use the `list()` function*.

```python
# create a list of words
sent = ["In", "the", "beginning"]

# convert it to a list of bigrams
list(bigrams(sent))
```

**Generating Random Text**:

The following program:
1. obtains all bigrams from the text of the book of Genesis
2. constructs a CFD to record which words are most likely to follow a given word
3. the generate_model() function uses this data, and a seed word, to generate random text.

```python
def generate_model(cfdist, word, num=15):
    """ Generate text based on bigrams."""
    
    for i in range(num):
        # print word with a space at the end
        print(word, end=" ")
        # reset word to the most likely token in that context
        word = cfdist[word].max()
        
# assign the words of genesis corpora to var text
text = nltk.corpus.genesis.words("english-kjv.txt")
# convert the list of words to bigrams
bigrams = nltk.bigrams(text)
# create a CFD for each bigram
cfd = nltk.ConditionalFreqDist(bigrams)

# generate words based
print(generate_model(cfd, 'living'))
# check the FreqDist of the word 
cfd['living']
```
