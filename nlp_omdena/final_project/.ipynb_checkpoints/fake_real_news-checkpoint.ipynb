# Text Preprocessing

```python
# define stopword list
stop_words = stopwords.words('english')
```

```python
def normalize_text(text):
    """Normalize the text with basic preprocessing tasks."""
    
    # clean text from html tags
    soup = BeautifulSoup(text, 'html.parser')
    # get clean text
    text = soup.get_text()
    # remove special characters
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    # remove trailing whitespace
    text = text.strip()
    # expand contractions
    text = contractions.fix(text)
    # tokenize text
    tokens = word_tokenize(text)
    # lower-case tokens
    tokens = [token.lower() for token in tokens]
    # remove stopwords
    tokens = [token for token in tokens if token not in stop_words]
    # pos-tag tokens
    tokens = pos_tag(tokens)
    # convert tags into wordnet format
    tag_map = {'j': wordnet.ADJ, 'v': wordnet.VERB, 'n': wordnet.NOUN, 'r': wordnet.ADV}
    tokens = [(word, tag_map.get(tag[0].lower(), wordnet.NOUN)) for word, tag in tokens]
    # instantiate WNL
    wnl = WordNetLemmatizer()
    # lemmatize tokens
    tokens = [wnl.lemmatize(word, tag) for word, tag in tokens]
    # re-create text from tokens
    text = ' '.join(tokens)
    
    return text
```

```python
# apply function on 'news' column
df['news_clean'] = df['news'].apply(normalize_text)

# print first 5 rorws
df.head()
```

```python
def get_corpus(text):
    """Tokenize a list of sentences."""
    tokens = []
    for sentence in text:
        for word in sentence.split():
            tokens.append(word.strip())
    return tokens
```

```python
# tokenize all news samples
corpus = get_corpus(df.news_clean)
# create a frequency distribution
fdist = FreqDist(corpus)
# plot the 10 most common words
fdist.plot(10);
```

```python
# instantiate Counter with news tokens
counter = Counter(corpus)
# extract the 10 most common words in the news
most_common = dict(counter.most_common(10))
# print the 10 most common words
most_common
```

```python
def get_top_text_ngrams(corpus, n, g):
    cv = CountVectorizer(ngram_range=(g, g)).fit(corpus)
    bow = cv.transform(corpus)
    sum_words = bow.sum(axis=0)
    words_freq = [(word, sum_words[0, idx]) for word, idx in cv.vocabulary_.items()]
    words_freq = sorted(words_freq, key = lambda x: x[1], reverse=True)
    return words_freq[:n]
```

```python
# unigram analysis
plt.figure(figsize=(16, 9))
most_common_uni = dict(get_top_text_ngrams(df.news_clean, 10, 1))
sns.barplot(x=list(most_common_uni.values()), y=list(most_common_uni.keys()));
```

```python
# bigram analysis
plt.figure(figsize=(16, 9))
most_common_bi = dict(get_top_text_ngrams(df.news_clean, 10, 2))
sns.barplot(x=list(most_common_bi.values()), y=list(most_common_bi.keys()));
```

```python
# trigram analysis
plt.figure(figsize=(16, 9))
most_common_tri = dict(get_top_text_ngrams(df.news_clean, 10, 3))
sns.barplot(x=list(most_common_tri.values()), y=list(most_common_tri.keys()));
```

```python
# split train and test data
X_train, X_test, y_train, y_test = train_test_split(df.news_clean, df.label, random_state=22)
# print shape
X_train.shape, y_train.shape
```

```python
# instantiate Tokenizer
tokenizer = text.Tokenizer(num_words=10_000, lower=False)
# fit train data on tokenizer
tokenizer.fit_on_texts(X_train)
# tokenize & vectorize train data
tokenized_train = tokenizer.texts_to_sequences(X_train)
# keep all news to 300 words and pad those with less than 300 
X_train = sequence.pad_sequences(tokenized_train, maxlen=300)
# tokenize & vectorize train data
tokenized_test = tokenizer.texts_to_sequences(X_test)
# keep all news to 300 words and pad those with less than 300 
X_test = sequence.pad_sequences(tokenized_test, maxlen=300)
```

```python
# print word mapping
tokenizer.word_index
```

We will use a pre-trained word vector which it can be found [here](https://www.kaggle.com/bertcarremans/glovetwitter27b100dtxt?select=glove.twitter.27B.100d.txt). 

```python
# import pre-trained word vector
EMBEDDING_FILE = r"C:\Users\spani\Documents\GitHub\nlp_resources\nlp_omdena\final_project\glove.twitter.27B.100d.txt"
```

```python
def get_coefs(word, *arr):
    return word, np.asarray(arr, dtype='float32')

embeddings_index = dict(get_coefs(*o.rstrip().rsplit(' ')) for o in open(EMBEDDING_FILE, encoding='utf8'))
```

```python
all_embs = np.stack(list(embeddings_index.values()))
emb_mean, emb_std = all_embs.mean(), all_embs.std()
embed_size = all_embs.shape[1]

word_index = tokenizer.word_index
nb_words = min(10_000, len(word_index))
embedding_matrix = np.random.normal(emb_mean, emb_std, (nb_words, embed_size))

for word, i in word_index.items():
    if i >= 10_000:
        continue
    embedding_vector = embeddings_index.get(word)
    if embedding_vector is not None:
        embedding_matrix[i] = embedding_vector    
```

# Building the Model

```python
# define callbacks
# batch_size = 256
# epochs = 10
# embed_size = 100
lr_reduction = ReduceLROnPlateau(monitor='val_accuracy', patience=2, verbose=1, factor=0.5, min_lr=0.00001)
es = EarlyStopping(monitor='val_loss', mode='min', patience=10, verbose=1)
```

```python
# build model
model = Sequential([
    # non-trainable embedding layer
    Embedding(10_000, output_dim=100, weights=[embedding_matrix], input_length=300, trainable=False),
    # LSTM layers
    LSTM(units=128, return_sequences=True, recurrent_dropout=0.25, dropout=0.25),
    LSTM(units=64, recurrent_dropout=0.1, dropout=0.1),
    # Dense layers
    Dense(units=32, activation='relu'),
    # output layer (sigmoid is used for binary classification problem)
    Dense(units=1, activation='sigmoid')
])

# model summary
model.summary()
```

```python
# compile model
model.compile(optimizer=Adam(learning_rate=0.01),
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
# train model
history = model.fit(X_train,
                    y_train,
                    batch_size=256,
                    validation_data = (X_test, y_test),
                    epochs=10,
                    callbacks=[lr_reduction, es])
```

```python
# evalute model
print("Accuracy of the model on Training Data is: " , model.evaluate(X_train,y_train)[1]*100 , "%")
print("Accuracy of the model on Testing Data is: " , model.evaluate(X_test,y_test)[1]*100 , "%")
```

```python
# visualize results
epochs = [i for i in range(10)]
fig , ax = plt.subplots(1,2)
train_acc = history.history['accuracy']
train_loss = history.history['loss']
val_acc = history.history['val_accuracy']
val_loss = history.history['val_loss']
fig.set_size_inches(20,10)

ax[0].plot(epochs , train_acc , 'go-' , label = 'Training Accuracy')
ax[0].plot(epochs , val_acc , 'ro-' , label = 'Testing Accuracy')
ax[0].set_title('Training & Testing Accuracy')
ax[0].legend()
ax[0].set_xlabel("Epochs")
ax[0].set_ylabel("Accuracy")

ax[1].plot(epochs , train_loss , 'go-' , label = 'Training Loss')
ax[1].plot(epochs , val_loss , 'ro-' , label = 'Testing Loss')
ax[1].set_title('Training & Testing Loss')
ax[1].legend()
ax[1].set_xlabel("Epochs")
ax[1].set_ylabel("Loss")
plt.show()
```
