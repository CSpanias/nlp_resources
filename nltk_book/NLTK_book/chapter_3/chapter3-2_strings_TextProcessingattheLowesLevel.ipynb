###  3.2.1 Basic Operations with Strings

#### Multi-line Strings

```python
# using backlash
msg = "This is " \
      "a multi-line" \
      " string."
print("Using backlash: {}\n".format(msg))

# using parentheses
msg_1 = ("This is"
         " a multi-line"
         " string.")
print("Using parentheses: {}\n".format(msg_1))

# using triple-quotes
msg_2 = """This is
        a multi-line
        string"""
print("Using triple-quotes: ", msg_2)
```

<a name="Slicing"></a>
###  3.2.2 Accessing Individual Characters

```python
from nltk.corpus import gutenberg

# get raw text (single string)
raw = gutenberg.raw('melville-moby_dick.txt')

# create a FreqDist while normalizing letters
fdist = nltk.FreqDist(ch.lower() for ch in raw if ch.isalpha())

# check the 5 most common letters
print(fdist.most_common(5), "\n")

# get the letters in descending order
print([char for (char, count) in fdist.most_common()])

# generate plot
fdist.plot();
```
